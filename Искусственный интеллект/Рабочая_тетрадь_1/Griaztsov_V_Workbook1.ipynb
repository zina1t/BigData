```python
import random
import matplotlib.pyplot as plt
x = np.array([random.random() for i in range(10)])
print(x)
print(np.median(x))
print(np.mean(x))
plt.scatter(range(len(x)), x)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Scatter Plot')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(1, 11, 1)
func = np.sqrt(1+np.exp(np.sqrt(x))+np.cos(x**2)) / np.abs(1-(np.sin(x)**3)) + np.log(np.abs(2*x))
plt.grid()
plt.plot(x, func)
y = x[:5]
a = func[:5]
plt.show()
plt.grid()
plt.scatter(y, a)
```

```python
import numpy as np
import matplotlib.pyplot as plt

from numpy import trapz
x = np.arange(1, 11, 1)
func = np.abs(np.cos(np.cos(x*np.e**(np.cos(np.cos(x)+np.log(np.log(x+1)))))))
plt.grid()
plt.plot(x, func)
plt.fill_between(x, func)
plt.show()
area = np.trapz(func)
print(area)
```

```python
x = int(input())
y = int(input())
print('1. x+y 2. x-y 3. x*y 4. x/y 5. e^(x+y) 6. sin(x+y) 7. cos(x+y) 8. x^y')
n = int(input())
match n:
    case 1:
        print(f'{x} + {y} = {x+y}')
    case 2:
        print(f'{x} - {y} = {x-y}')
    case 3:
        print(f'{x} * {y} = {x*y}')
    case 4:
        if y != 0:
            print(f'{x} ÷ {y} = {x/y}')
        else:
            print(x, y)
            print('На ноль делить нельзя')
    case 5:
        print(f'e^({x} + {y}) = {np.e**(x+y)}')
    case 6:
        print(f'sin({x} + {y}) = {np.sin(x+y)}')
    case 7:
        print(f'cos({x} + {y}) = {np.cos(x+y)}')
    case 8:
        print(f'{x}^{y} = {x**y}')
```
